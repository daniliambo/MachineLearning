### Индексирование

```python
# Slicing
normal_distribution = np.random.randn(6, 8)
print(normal_distribution)
# normal_distribution[1, 1]
# normal_distribution[:, 1] # 2nd column
# normal_distribution[::-1, 1]
# normal_distribution[::2, 1]
# normal_distribution[1, :]
# normal_distribution[1]
#
# # Берем значения по маске
# normal_distribution[(normal_distribution > 0)] # 1d!!!
# normal_distribution[np.random.randn(6) > 0] # 2d!!!
normal_distribution[:, np.random.randn(8) > 0] # 2d!!!
#
# # Берем определенные индексы
# ids = np.arange(0, 6, 2)
# normal_distribution[ids]
```

```python
# Operations with slices
normal_distribution = np.random.randn(6, 4)
print(normal_distribution)
normal_distribution[:, 1] = 10
print(normal_distribution)
```

### Математические операции

```python
# Среднее
a = np.random.randn(8, 8)
print(a)
print(np.mean(a))
print(a.mean())
print(a.mean(-1))
```

```python
# Максимум
a = np.random.randn(8, 8)
print(a)
print(np.max(a))
print(a.max())
print(a.max(-1)) # max из каждого array
```

```python
# Argmax
a = np.random.randn(8, 8)
print(a)
print(np.argmax(a)) # Position in flattened array
print(a.argmax(-1)) # max indices в каждом array
```

```python
# Get unique elements
np.unique(np.random.randint(1, 100, 1000), return_counts=True)
```

```python
np_vector = np.random.randn(16)
np_matrix = np.random.randn(16, 16)

# Elementvise operations
np_vector * np_vector
np_vector - np_vector
np_vector + np_vector
np_vector / np_vector
np_matrix * np_vector
np_matrix - np_vector
np_matrix + np_vector
np_matrix / np_vector

# Tensor operations
np_vector.dot(np_vector) # Production
np_matrix.T # Transpose
```

```python
a = np.random.randn(8, 8)
print(np.linalg.norm(a))
print(np.linalg.norm(a, axis=-1))
```

### Изменение размерности

```python
# Shape
a = np.linspace(0, 1, 12)
print(a)
a = a.reshape((2, 6))
print(a)
a = a.reshape((-1, 3))
print(a)
```

```python
a = np.random.randn(4, 4)
b = np.random.randn(4, 4, 4)
a = np.expand_dims(a, 1)
a = np.broadcast_to(a, b.shape)
print(a)
```
